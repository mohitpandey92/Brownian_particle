```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import linalg as LA
```

### Dead boundary condition

```python
def Ham_dbc_minus(M):
    T = -(np.diag(np.ones(shape=(M-1,)), -1) + np.diag(np.ones(shape=(M-1,)),  1))# DBC
    H=T
    return H

def Ham_dbc_plus(M):
    T = (np.diag(np.ones(shape=(M-1,)), -1) + np.diag(np.ones(shape=(M-1,)),  1))# DBC
    H=T
    return H
```

```python
print Ham_dbc_minus(5)
print Ham_dbc_plus(5)
```

```python
M=100
w_dbc_plus,v_dbc_plus=LA.eigh(Ham_dbc_plus(M))
w_dbc_minus,v_dbc_minus=LA.eigh(Ham_dbc_minus(M))
```

### Periodic boundary condition: eigenstates ( N even)

```python
def Ham_pbc_minus(M):
    T = -(np.diag(np.ones(shape=(M-1,)), -1) + np.diag(np.ones(shape=(M-1,)),  1))
    T[0][M-1]=-1.0 # PBC
    T[M-1][0]=-1.0 
    H=T
    return H

def Ham_pbc_plus(M):
    T = (np.diag(np.ones(shape=(M-1,)), -1) + np.diag(np.ones(shape=(M-1,)),  1))
    T[0][M-1]=1.0 # PBC
    T[M-1][0]=1.0
    H=T
    return H
```

```python
M=100
w_pbc_minus,v_pbc_minus=LA.eigh(Ham_pbc_minus(M))
w_pbc_plus,v_pbc_plus=LA.eigh(Ham_pbc_plus(M))
```

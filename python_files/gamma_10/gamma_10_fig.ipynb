```python
def fn(t):
    D=10.0
    omega=1.0/10.0
    return 25*np.exp(-2*omega*t)+ (D/omega)*(1- np.exp(-2*omega*t) )
def fn1(t):
    D=10.0
    omega=2.0/10.0
    return 25.00*np.exp(-2*omega*(t))+ (D/omega)*(1- np.exp(-2*omega*(t))) 
```
